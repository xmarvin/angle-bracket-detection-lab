```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
%matplotlib inline 
```

```python
image = cv2.imread('sample6-highlighted.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
def show_countors(imgray):
    ret, thresh = cv2.threshold(imgray, 127, 255, 0)
    _, contours, _  = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    fig,ax = plt.subplots(1)
    ax.imshow(imgray)
    for cords in contours:
         x = [a[0] for a in np.squeeze(cords)]
         y = [a[1] for a in np.squeeze(cords)]
         minx, maxx = np.min(x), np.max(x)
         miny, maxy = np.min(y), np.max(y)
         rect = Rectangle((minx,miny),maxx-minx,maxy-miny,linewidth=1,edgecolor='r',facecolor='none')
         ax.add_patch(rect)
```

```python
show_countors(image)
```

```python
show_countors(cv2.imread('mask_bracket/data/higlighted.jpg'))
```
