```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.patches import Rectangle
%matplotlib inline 
```

```python
image = cv2.imread('sample6-highlighted.jpg')
```

```python
imgray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(imgray, 127, 255, 0)
_, contours, _  = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
fig,ax = plt.subplots(1)
ax.imshow(image)
for cords in contours:
     x = [a[0] for a in np.squeeze(cords)]
     y = [a[1] for a in np.squeeze(cords)]
     minx, maxx = np.min(x), np.max(x)
     miny, maxy = np.min(y), np.max(y)
     rect = Rectangle((minx,miny),maxx-minx,maxy-miny,linewidth=1,edgecolor='r',facecolor='none')
     ax.add_patch(rect)
```
